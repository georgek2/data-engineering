```python
if response.status_code == 200:
    data = response.json() # Grab the json data
    # Convert the asteroid data to a Pandas DataFrame
    asteroids_df = pd.json_normalize(data["near_earth_objects"].values())
    neos = data['near_earth_objects']

    print(neos.keys())
    print(neos.values())
    print(len(neos.values()))
    print(neos['2024-03-05'][0].keys())
```

```python
id = []
name = []
absolute_magnitude_h = []
ed_km_min = []
ed_km_max = []
ed_miles_min = []
ed_miles_max = []
ed_ft_min = []
ed_ft_max = []
hazardous = []

for i, key in enumerate(neos.keys()):
    id.append(neos[key][i]['id'])
    name.append(neos[key][i]['name'])
    absolute_magnitude_h.append(neos[key][i]['absolute_magnitude_h'])
    ed_km_min.append(neos[key][i]['estimated_diameter']['kilometers']['estimated_diameter_min'])
    ed_km_max.append(neos[key][i]['estimated_diameter']['kilometers']['estimated_diameter_max'])
    ed_miles_min.append(neos[key][i]['estimated_diameter']['miles']['estimated_diameter_min'])
    ed_miles_max.append(neos[key][i]['estimated_diameter']['miles']['estimated_diameter_max'])
    ed_ft_min.append(neos[key][i]['estimated_diameter']['feet']['estimated_diameter_min'])
    ed_ft_max.append(neos[key][i]['estimated_diameter']['feet']['estimated_diameter_max'])
    hazardous.append(neos[key][i]['is_potentially_hazardous_asteroid'])
```

```python
data_obj = {
    'id': id,
    'name': name,
    'absolute_magnitude_h': absolute_magnitude_h,
    'ed_km_min': ed_km_min,
    'ed_km_max': ed_km_max,
    'ed_miles_min': ed_miles_min,
    'ed_miles_max': ed_miles_max,
    'ed_ft_min': ed_ft_min,
    'ed_ft_max': ed_ft_max,
    'hazardous': hazardous
}

df = pd.DataFrame(data_obj)
df
```

```python
# Feature Engineering
# Calculating mean diameters: km, miles, feet

df['ed_km_mean'] = df.apply(lambda x: (x['ed_km_min'] + x['ed_km_max'])/2, axis=1)
df['ed_miles_mean'] = df.apply(lambda x: (x['ed_miles_min'] + x['ed_miles_max'])/2, axis=1)
df['ed_ft_mean'] = df.apply(lambda x: (x['ed_ft_min'] + x['ed_ft_max'])/2, axis=1) 
```
